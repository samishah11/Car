```python
import cv2

# Load pre-trained car classifier
car_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_car.xml')

# Function to detect cars in an image and count them
def detect_and_count_cars(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Detect cars
    cars = car_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5)
    
    # Draw rectangles around the detected cars and count them
    car_count = 0
    for (x, y, w, h) in cars:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        car_count += 1
    
    return image, car_count

# Read input video or camera stream
cap = cv2.VideoCapture('video3.mp4')  # Replace 'input_video.mp4' with 0 for camera stream

# Initialize car count
total_cars = 0

# Loop through frames and detect cars
while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Detect cars in the current frame and count them
    frame_with_cars, car_count = detect_and_count_cars(frame)
    
    # Increment total car count
    total_cars += 1
    
    # Display the frame with cars and car count
    cv2.putText(frame_with_cars, 'Total Cars: {}'.format(total_cars), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv2.imshow('Car Detection', frame_with_cars)
    
    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```
